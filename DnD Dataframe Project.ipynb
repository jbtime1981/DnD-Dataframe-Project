1) Built two dataframes using pd.DataFrame()
2) Used .describe() and .info()
3) Converted data types from *obj/str* to *int*

5) Pulled meaningful insights on player character builds

```python
# Converting select objects to integers

com_dnd_df['Level'] = com_dnd_df['Level'].astype(str).astype(int)

com_dnd_df['Armor Class'] = com_dnd_df['Armor Class'].astype(str).astype(int)

com_dnd_df['HP Max'] = com_dnd_df['HP Max'].astype(str).astype(int)

com_dnd_df['EXP'] = com_dnd_df['EXP'].astype(str).astype(int)

com_dnd_df.info()
```

```python
# Sorting the data according to HP Max column for bar chart

df_sorted_hp = com_dnd_df.sort_values('HP Max')
df_sorted_hp
```

```python
# Bar chart showing Player and HP Max

plt.bar(x=df_sorted_hp['Player Name'], height=df_sorted_hp['HP Max'], label = "Player's Health Points Max", color='g')
plt.plot()

plt.xlabel('Player Name')
plt.ylabel('HP Max')
plt.title("Player's Max Health Points")
plt.legend()
plt.xticks(rotation=90)

plt.show()
```

#### Insights on Health Points Chart

Pierce's character has the most health points, and Abed has the least.  This means that Pierce can take the most damage without dying, and Abed cannot take much damage at all by comparison.   

```python
# Sorting the data according to Armor Class column for bar chart

df_sorted_ac = com_dnd_df.sort_values('Armor Class')
df_sorted_ac
```

```python
# Bar chart showing Player and Armor Class

plt.bar(x=df_sorted_ac['Player Name'], height=df_sorted_ac['Armor Class'], label = "Player's Armor Class", color='r')
plt.plot()

plt.xlabel('Player Name')
plt.ylabel('Armor Class')
plt.title("Player's Armor Class")
plt.legend()
plt.xticks(rotation=90)

plt.show()
```

#### Insights on Armor Class Chart

Pierce, Troy, and Shirley are the players with the highest armor classes, making them ideal tanks.  Abed's character has the lowest armor class, making him best as a ranged player (or at least one protected by the group).
